```python
#script observation method
def query_obs(target=str, exptime=float, offset=float, num_start=1, num_end=float):
    obs.use_query(target=target, exptime=exptime, offset=offset, num_start=num_start, num_end=num_end)

def coord_obs(target=str, ra=float, dec=float, exptime=float, offset=float, num_start=1, num_end=float):
    obs.use_coord(target, ra=ra, dec=dec, exptime=exptime, offset=offset, num_start=num_start, num_end=num_end)
```

```python
#calibration frame exposure
def calibration_frame(iter=int, darktime=list, overwirte=False):
    obs.calib(num_iter=iter, darktime=darktime, overwirte=overwirte)
```
